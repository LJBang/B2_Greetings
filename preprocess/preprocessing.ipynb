```python
import kss

sentence_tokenized_text = []
for i, line in enumerate(lines):
    if i > 100:     # 전체 wikipedia 문서는 사이즈가 크므로, 일부만 테스트.
        break
    line = line.strip()
    for sent in kss.split_sentences(line):
        sentence_tokenized_text.append(sent.strip())
```

```python
for i in range(10):
    print(sentence_tokenized_text[i])
```

```python
punct = "/-'?!.,#$%\'()*+-/:;<=>@[\\]^_`{|}~" + '""“”’' + '∞θ÷α•à−β∅³π‘₹´°£€\×™√²—–&'
punct_mapping = {"‘": "'", "₹": "e", "´": "'", "°": "", "€": "e", "™": "tm", "√": " sqrt ", "×": "x", "²": "2", "—": "-", "–": "-", "’": "'", "_": "-", "`": "'", '“': '"', '”': '"', '“': '"', "£": "e", '∞': 'infinity', 'θ': 'theta', '÷': '/', 'α': 'alpha', '•': '.', 'à': 'a', '−': '-', 'β': 'beta', '∅': '', '³': '3', 'π': 'pi', }
```

```python
def clean_punc(text, punct, mapping):
    for p in mapping:
        text = text.replace(p, mapping[p])
    
    for p in punct:
        text = text.replace(p, f' {p} ')
    
    specials = {'\u200b': ' ', '…': ' ... ', '\ufeff': '', 'करना': '', 'है': ''}
    for s in specials:
        text = text.replace(s, specials[s])
    
    return text.strip()
```

```python
cleaned_corpus = []
for sent in sentence_tokenized_text:
    cleaned_corpus.append(clean_punc(sent, punct, punct_mapping))
```

```python
for i in range(0, 10):
    print(cleaned_corpus[i])
```

```python
import re

def clean_text(texts):
    corpus = []
    for i in range(0, len(texts)):
        review = str(texts[i])
        # review = re.sub(r'\d+','', str(texts[i]))# 숫자 제거
        # review = review.lower() # 소문자
        review = re.sub(r'<[^>]+>','',review) # HTML 태그 제거
        review = re.sub(r'\s+', ' ', review) # 두 칸 이상 여백 제거
        review = re.sub(r"^\s+", '', review) # 시작 여백 제거
        review = re.sub(r'\s+$', '', review) # 마지막 여백 제거
        corpus.append(review)
    return corpus
```

```python
basic_preprocessed_corpus = clean_text(cleaned_corpus)
```

```python
for i in range(0, 10):
    print(basic_preprocessed_corpus[i])
```
